```python
#Import libraries
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset = pd.read_csv('inaccuracy4.csv')

#Empty lists 
FAMAS = ["Famas"] #0
AUG_unscoped = ["AUG_unscoped"] #1
AUG_scoped = ["AUG_scoped"] #2
M4A1S = ["M4A1S"] #3
M4A4 = ["M4A4"] #4
Ak47 = ["Ak47"] #5
Galil = ["Galil"] #6
SG556_unscoped = ["SG556_unscoped"] #7
SG556_scoped_ = ["SG556_scoped_"] #8
MP5SD = ["MP5SD"] #9
MAC10 = ["MAC10"] #10
MP9 = ["MP9"] #11
MP7 = ["MP7"] #12
UMP = ["UMP"] #13
P90 = ["P90"] #14
PPBIZON = ["PPBIZON"] #15
CZ75 = ["CZ75"] #16
WEAPONS = ['WEAPONS']
for x in range(1,100):
    WEAPONS.append(x)


file_to_output = open('inaccuracyfg.csv',mode = 'w', newline= '')
csv_writer = csv.writer(file_to_output,delimiter = ',')

#Accurate distance defined as 100% of shots landing within a 30cm(0.3m) cirlce
#Therefore, the accuracy is modelled as such: at a given distance, a round has an equal chance of landing on any pixel in a circle of diameter 30cm
#d = 2r
#deviation_diameter = 0.3
deviationRadius = 0.15

#Head of size, d = 30cm
#Circle = 0.25(pi)d^2
headDiameter = 0.3
sizeOfHead = 0.25 * np.pi * headDiameter * headDiameter

#distance from 25m
distance = 25.7


x = dataset.iloc[:,:]
for num in range(17):
    for num1 in range(100):
        if float(x.iloc[num1,num]) > 0:
            accurateDistance = dataset.iloc[num1,num]
            ratio = deviationRadius/accurateDistance
            absoluteRadius = distance * ratio
            absoluteSizeOfDeviation = np.pi * absoluteRadius * absoluteRadius
            chanceOfHittingHead = sizeOfHead / absoluteSizeOfDeviation            
            if chanceOfHittingHead > 1:
                chanceOfHittingHead = 1
            if num == 0:
                FAMAS.append(chanceOfHittingHead)
            if num == 1:
                AUG_unscoped.append(chanceOfHittingHead)
            if num == 2:
                AUG_scoped.append(chanceOfHittingHead)
            if num == 3 :
                M4A1S.append(chanceOfHittingHead)
            if num == 4 :
                M4A4.append(chanceOfHittingHead)                
            if num == 5 :
                Ak47.append(chanceOfHittingHead)                
            if num == 6 :
                Galil.append(chanceOfHittingHead)  
            if num == 7 :
                SG556_unscoped.append(chanceOfHittingHead)
            if num == 8 :
                SG556_scoped_.append(chanceOfHittingHead)                             
            if num == 9 :
                MP5SD.append(chanceOfHittingHead)
            if num == 10 :
                MAC10.append(chanceOfHittingHead)                             
            if num == 11 :
                MP9.append(chanceOfHittingHead)
            if num == 12 :
                MP7.append(chanceOfHittingHead)
            if num == 13 :
                UMP.append(chanceOfHittingHead)
            if num == 14 :
                P90.append(chanceOfHittingHead)                             
            if num == 15 :
                PPBIZON.append(chanceOfHittingHead)                             
            if num == 16 :
                CZ75.append(chanceOfHittingHead)

csv_writer.writerow(WEAPONS)
csv_writer.writerow(FAMAS)
csv_writer.writerow(AUG_unscoped)
csv_writer.writerow(AUG_scoped)
csv_writer.writerow(M4A1S)
csv_writer.writerow(M4A4)
csv_writer.writerow(Ak47)
csv_writer.writerow(Galil)
csv_writer.writerow(SG556_unscoped)
csv_writer.writerow(SG556_scoped_)
csv_writer.writerow(MP5SD)
csv_writer.writerow(MAC10)
csv_writer.writerow(MP9)
csv_writer.writerow(MP7)
csv_writer.writerow(UMP)
csv_writer.writerow(P90)
csv_writer.writerow(PPBIZON)
csv_writer.writerow(CZ75)
file_to_output.close()
```
